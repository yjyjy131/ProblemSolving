```python
from collections import deque


# 구슬 파괴
def destroy_marvel(d, s):
    for dv in range(1, s+1):
        nx = shark[0] + dir_dx[d] * dv
        ny = shark[1] + dir_dy[d] * dv

        if 0 <= nx < n and 0 <= ny < n:
            board[nx][ny] = 0


# 구슬 이동
def move_marvel():
    m_values = [0 for _ in range(n*n)]
    m_values[0] = board[shark[0]][shark[1]]

    m_idx = 0
    curr_dir = 1
    next_dir = {3: 2, 2: 4, 4: 1, 1: 3}
    curr_x, curr_y = shark[0], shark[1]

    # 구슬의 값들만 차례로 가져옴
    for go_size in range(1, n):
        for _ in range(2):
            curr_dir = next_dir.get(curr_dir)

            for _ in range(go_size):
                curr_x += dir_dx[curr_dir]
                curr_y += dir_dy[curr_dir]

                if board[curr_x][curr_y] != 0:
                    m_idx += 1
                    m_values[m_idx] = board[curr_x][curr_y]

    curr_dir = next_dir.get(curr_dir)
    for _ in range(n-1):
        curr_x += dir_dx[curr_dir]
        curr_y += dir_dy[curr_dir]
        if board[curr_x][curr_y] != 0:
            m_idx += 1
            m_values[m_idx] = board[curr_x][curr_y]

    # 구슬 재배치
    m_idx = 0
    curr_dir = 1
    curr_x, curr_y = shark[0], shark[1]
    new_board = [[0 for _ in range(n)] for _ in range(n)]
    new_board[curr_x][curr_y] = m_values[0]

    for go_size in range(1, n):
        for _ in range(2):
            curr_dir = next_dir.get(curr_dir)

            for _ in range(go_size):
                m_idx += 1
                curr_x += dir_dx[curr_dir]
                curr_y += dir_dy[curr_dir]
                new_board[curr_x][curr_y] = m_values[m_idx]

    curr_dir = next_dir.get(curr_dir)
    for _ in range(n):
        m_idx += 1
        if m_idx == len(m_values):
            break
        curr_x += dir_dx[curr_dir]
        curr_y += dir_dy[curr_dir]
        new_board[curr_x][curr_y] = m_values[m_idx]

    return new_board


# 구슬 폭발
def remove_marvel():
    curr_dir = 1
    next_dir = {3: 2, 2: 4, 4: 1, 1: 3}
    curr_x, curr_y = shark[0], shark[1]

    dq = deque([(curr_x, curr_y)])
    prev_m = board[curr_x][curr_y]
    rm_list = []

    for go_size in range(1, n):
        for _ in range(2):
            curr_dir = next_dir.get(curr_dir)

            for _ in range(go_size):
                curr_x += dir_dx[curr_dir]
                curr_y += dir_dy[curr_dir]
                curr_m = board[curr_x][curr_y]
                if prev_m != curr_m:
                    if len(dq) >= 4:
                        while dq:
                            cx, cy = dq.popleft()
                            cnt[prev_m] += 1
                            rm_list.append((cx, cy))

                    prev_m = curr_m
                    dq = deque([(curr_x, curr_y)])
                else:
                    dq.append((curr_x, curr_y))

    curr_dir = next_dir.get(curr_dir)
    for _ in range(n-1):
        curr_x += dir_dx[curr_dir]
        curr_y += dir_dy[curr_dir]
        curr_m = board[curr_x][curr_y]
        if prev_m != curr_m:
            if len(dq) >= 4:
                while dq:
                    cx, cy = dq.popleft()
                    cnt[prev_m] += 1
                    rm_list.append((cx, cy))

            prev_m = curr_m
            dq = deque([(curr_x, curr_y)])
        else:
            dq.append((curr_x, curr_y))

    return rm_list


# 구슬 변화
def change_marvel():
    curr_dir = 1
    next_dir = {3: 2, 2: 4, 4: 1, 1: 3}
    curr_x, curr_y = shark[0], shark[1]

    prev_m = board[curr_x][curr_y]
    cnt_m = 1
    groups = []

    for go_size in range(1, n):
        for _ in range(2):
            curr_dir = next_dir.get(curr_dir)

            for _ in range(go_size):
                curr_x += dir_dx[curr_dir]
                curr_y += dir_dy[curr_dir]
                curr_m = board[curr_x][curr_y]
                if prev_m != curr_m:
                    groups.append(cnt_m)
                    groups.append(prev_m)
                    prev_m = curr_m
                    cnt_m = 1
                else:
                    cnt_m += 1

    groups = groups[2:n*n+2]

    # 구슬 재배치
    m_idx = -1
    curr_dir = 1
    curr_x, curr_y = shark[0], shark[1]
    new_board = [[0 for _ in range(n)] for _ in range(n)]
    new_board[curr_x][curr_y] = -1

    for go_size in range(1, n):
        for _ in range(2):
            curr_dir = next_dir.get(curr_dir)
            for _ in range(go_size):
                m_idx += 1
                if m_idx == len(groups):
                    return new_board
                curr_x += dir_dx[curr_dir]
                curr_y += dir_dy[curr_dir]
                new_board[curr_x][curr_y] = groups[m_idx]

    curr_dir = next_dir.get(curr_dir)
    for _ in range(n-1):
        m_idx += 1
        if m_idx == len(groups):
            return new_board
        curr_x += dir_dx[curr_dir]
        curr_y += dir_dy[curr_dir]
        new_board[curr_x][curr_y] = groups[m_idx]

    return new_board


if __name__ == '__main__':
    n, m = map(int, input().split())
    board = [list(map(int, input().split())) for _ in range(n)]
    magic_list = [list(map(int, input().split())) for _ in range(m)]

    dir_dx = [None, -1, 1, 0, 0]
    dir_dy = [None, 0, 0, -1, 1]
    shark = (n//2, n//2)
    board[shark[0]][shark[1]] = -1

    cnt = [0, 0, 0, 0]

    for mi in range(m):
        # 얼음 파편 구슬 파괴
        curr_d, curr_s = magic_list[mi]
        destroy_marvel(curr_d, curr_s)

        # 구슬 이동
        board = move_marvel()

        # 폭발할 구슬이 있는지 체크 후 반복
        while 1:
            # 구슬 폭발
            rm_target = remove_marvel()

            if not rm_target:
                break

            for rx, ry in rm_target:
                board[rx][ry] = 0

            # 구슬 이동
            board = move_marvel()

        # 구슬 변화
        board = change_marvel()

    print(cnt[1] + 2 * cnt[2] + 3 * cnt[3])




```
