```python
go = {'u':(0, 1), 'd':(0, -1), 'l':(-1, 0), 'r':(1, 0)}
```

```python
down = {'u':(0, -1), 'd':(0, 1), 'l':(1, 0), 'r':(-1, 0)}
```

```python
l_dir = { 'u':'l', 'l':'d', 'd':'r', 'r':'u'}
```

```python
r_dir = { 'u':'r', 'l':'u', 'd':'l', 'r':'d'}
```

```python
for i in range(n) :
  d = 'u'
  x = [0]
  y = [0]
  curr = [0, 0]

  for r in route[i] :
    if r == 'L' :
      d = l_dir[d]

    elif r == 'R' :
      d = r_dir[d]

    elif r == 'F' :
      curr[0] += go[d][0]
      curr[1] += go[d][1]
      x.append(curr[0])
      y.append(curr[1])

    elif r == 'B' :
      curr[0] += down[d][0]
      curr[1] += down[d][1]
      x.append(curr[0])
      y.append(curr[1])


  max_x = max(x) - min(x)
  max_y = max(y) - min(y)
  print(max_x * max_y)    
```
